```python
%load_ext autoreload
%autoreload 2

import keras
from keras.models import load_model

from data_loader import get_FMNIST, get_cifar10
from visualizer import compare_FMNIST, compare_CIFAR
from ae_models import get_CAE_FMNIST_Model, get_CAE_CIFAR_Model,  get_AE_FMNIST_model, get_AE_CIFAR_model, train_model
```

```python
(x_train, x_test, labels_test) = get_FMNIST()
(x_train_c, x_test_c, labels_test_c) = get_cifar10()
```
